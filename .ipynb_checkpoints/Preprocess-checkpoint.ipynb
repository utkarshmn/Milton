```python
import pandas as pd
import re
from datetime import datetime
import urllib.request
import json
def pretty(obj):
    return json.dumps(obj, sort_keys=True, indent=2)
```

```python
press = pd.read_csv('Data/FOMCPressLinks.txt', sep=" ", header=None, names=['Link'])
press['Link'] = press['Link'].apply(str)

test = press.iloc[66, 0]
print(test)
strDtFromLink = lambda x: re.findall(r"monetary\.?([ \d.]+)", x, re.IGNORECASE | re.MULTILINE)[0]

dateFromLink = lambda x: datetime.strptime(re.findall(r"monetary\.?([ \d.]+)", x, re.IGNORECASE | re.MULTILINE)[0],
                                           '%Y%m%d').date()
print(dateFromLink(test))

press['MeetingDate'] = [dateFromLink(x) for x in press['Link']]
press = press.iloc[:,::-1].set_index('MeetingDate').sort_index()
print(press)
```

```python
#test
for idx, row in press.iterrows():
    print(idx, strDtFromLink(row['Link']))
```

```python


from bs4 import BeautifulSoup, SoupStrainer
from bs4.element import Comment
import urllib.request

html = urllib.request.urlopen(test).read()
```

```python
parse_only = SoupStrainer(id=['content'])
soup = BeautifulSoup(html, 'html.parser', parse_only=parse_only)
divs = soup.find_all("div",{"class":"col-xs-12 col-sm-8 col-md-8"})
date = soup.find_all("p", {"class":"article__time"})
def scrapeFPR(urlLink):
    #Scrapes FOMC meeting page returning a list with [date, content]
    html = urllib.request.urlopen(urlLink).read()
    # don't waste time reading anything except the main contents
    parse_only = SoupStrainer(id=['content'])
    soup = BeautifulSoup(html, 'html.parser', parse_only=parse_only)
    
    # find main body and date for presser
    divs = soup.find_all("div",{"class":"col-xs-12 col-sm-8 col-md-8"})
    date = soup.find_all("p", {"class":"article__time"})
    numPgs = len(divs)
    meat = ''
    for each in divs:
        meat += str(each.get_text())
    meat = meat.replace('\n', ' ').replace('\r', '')
    import dateutil.parser
    date = str([x.get_text() for x in date][0])
    date = dateutil.parser.parse(date).date()
    return [date, meat]
        
errors = []
for idx, row in press.iterrows():
    res = scrapeFPR(row['Link'])
    press.loc[idx,'RelDate'] = res[0]
    press.loc[idx,'Presser'] = res[1]
    print (res[0]==idx)
    print (str(idx)+' saved')
```
